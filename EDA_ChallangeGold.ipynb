#Import Library dan Data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/data.csv", encoding = "latin")
df
```

#Explore Data

```python
df.info()
```

```python
df.describe()
```

```python
## menghitung jumlah kata
def split_sentences(s):
  return len(s.split())

df['total_kata'] = df['Tweet'].apply(split_sentences)
df
```

## Sum

```python
sum_HS = df['total_kata'].groupby(df['HS']).sum()
sum_HS = sum_HS.to_list()
sum_HS
```

```python
sum_ABS = df['total_kata'].groupby(df['Abusive']).sum()
sum_ABS = sum_ABS.to_list()
sum_ABS
```

```python
plotdata = pd.DataFrame({
    "HS":sum_HS,
    "Abusive":sum_ABS
    }, 
    index=["0", "1"]
)
plotdata.plot(kind="bar")
plt.title("Perbandingan jumlah kata yang mengandung HS atau Abusive")
plt.xlabel("Label")
plt.ylabel("Jumlah kata")
```

## Mean

```python
# mencari mean dari label Hate Speech
mean_HS = df['total_kata'].groupby(df['HS']).mean()
mean_HS = mean_HS.to_list()
mean_HS
```

```python
# mencari mean dari label Abusive
mean_ABS = df['total_kata'].groupby(df['HS']).mean()
mean_ABS = mean_ABS.to_list()
mean_ABS
```

```python
plotdata = pd.DataFrame({
    "HS":mean_HS,
    "Abusive":mean_ABS
    }, 
    index=["0", "1"]
)
plotdata.plot(kind="bar")
plt.title("Perbandingan mean dari data yang mengandung HS atau Abusive")
plt.xlabel("Label")
plt.ylabel("Jumlah kata")
```

```python
# check jumlah data yang mengandung ujaran kebencian
temp_HS = df.groupby('HS').count()['Tweet'].reset_index().sort_values(by='Tweet',ascending=False)
temp_HS.style.background_gradient(cmap='Purples')
temp_HS
```

```python
explode = (0.1, 0)

plt.pie(temp_HS['Tweet'],labels=['NOT HS','HS'], startangle=90, autopct='%1.1f%%', shadow=True, explode = explode)
plt.axis('equal')
plt.title("Perbandingan Jumlah Data yang Mengandung Ujaran Kebencian")
plt.legend()
plt.show()
```
